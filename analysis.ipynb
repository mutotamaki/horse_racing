```python
# 有馬記念の結果
result = ["ドウデュース", "スターズオンアース", "タイトルホルダー"]

# 有馬記念の予想リスト
predictions = [
    ["ジャスティンパレス", "ディープボンド", "ヒートオンビート"],
    ["タスティエーラ", "スターズオンアース", "ソールオリエンス"],
    ["ドウデュース", "ディープボンド", "タイトルホルダー"],
    ["タスティエーラ", "タイトルホルダー", "ソールオリエンス"],
    ["ジャスティンパレス", "ドウデュース", "ソールオリエンス"],
    ["ディアスティマ", "スターズオンアース", "ウインマリリン"],
    ["タスティエーラ", "タイトルホルダー", "ドウデュース"],
    ["ディープボンド", "ホウオウエミーズ", "ジャスティンパレス"]
]

# 1着、2着、3着それぞれの一致数を初期化
first_place_matches = 0
second_place_matches = 0
third_place_matches = 0

# 各予想をチェックし、一致する場合は対応する一致数を増やす
for prediction in predictions:
    if prediction[0] == result[0]:
        first_place_matches += 1
    if prediction[1] == result[1]:
        second_place_matches += 1
    if prediction[2] == result[2]:
        third_place_matches += 1

# 一致率を計算
total_predictions = len(predictions)
first_place_accuracy = first_place_matches / total_predictions
second_place_accuracy = second_place_matches / total_predictions
third_place_accuracy = third_place_matches / total_predictions

# 結果を出力
print(f"1着の一致率: {first_place_accuracy * 100}%")
print(f"2着の一致率: {second_place_accuracy * 100}%")
print(f"3着の一致率: {third_place_accuracy * 100}%")
```
